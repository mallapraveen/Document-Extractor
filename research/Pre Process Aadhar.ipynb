# Address Generation

```python
addr = pd.read_csv('./Database/temp_add.csv',)
addr['len'] = addr['add'].str.split(',').apply(len)

addr1 = []
addr2 = []
addr3 = []
addr4 = []
addr5 = []

for i in range(len(addr)):
    add1 = ''
    add2 = ''
    add3 = ''
    add4 = ''
    add5 = ''
    temp = addr['add'][i].split(',')
    for i in range(len(temp)):
        if(i in (0,1)):
            add1+=','
            add1+=temp[i].strip()
        elif(i in (2,3)):
            add2+=','
            add2+=temp[i].strip()
        elif(i in (4,5)):
            add3+=','
            add3+=temp[i].strip()
        elif(i in (6,7)):
            add4+=','
            add4+=temp[i].strip()
        else:
            add5+=','
            add5+=temp[i].strip()
    addr1.append(add1[1:])
    addr2.append(add2[1:])
    addr3.append(add3[1:])
    addr4.append(add4[1:])
    addr5.append(add5[1:-1])

pd.DataFrame({'add1':addr1,'add2':addr2,'add3':addr3,'add4':addr4,'add5':addr5}).to_csv('final_add.csv')
```

# Getting the Co ordinates of Image

```python
def click_event(event, x, y, flags, params):
 
    # checking for left mouse clicks
    if event == cv2.EVENT_LBUTTONDOWN:
 
        # displaying the coordinates
        # on the Shell
        print(x, ' ', y)
 
        # displaying the coordinates
        # on the image window
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, str(x) + ',' +
                    str(y), (x,y), font,
                    1, (255, 0, 0), 2)
        cv2.imshow('image', img)
 
    # checking for right mouse clicks    
    if event==cv2.EVENT_RBUTTONDOWN:
 
        # displaying the coordinates
        # on the Shell
        print(x, ' ', y)
 
        # displaying the coordinates
        # on the image window
        font = cv2.FONT_HERSHEY_SIMPLEX
        b = img[y, x, 0]
        g = img[y, x, 1]
        r = img[y, x, 2]
        cv2.putText(img, str(b) + ',' +
                    str(g) + ',' + str(r),
                    (x,y), font, 1,
                    (255, 255, 0), 2)
        cv2.imshow('image', img)
        
# reading the image
img = cv2.imread('./Database/Template/Front_Big.jpg', 1)


# displaying the image
cv2.imshow('image', img)
#cv2.resize(img,(780,540))
# setting mouse handler for the image
# and calling the click_event() function
cv2.setMouseCallback('image', click_event)

# wait for a key to be pressed to exit
cv2.waitKey(0)

# close the window
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import pytesseract
import re

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

img3 = cv2.imread(r'\ouput\Aadhar\Front_Long\Front_Long_15.JPG')   #any picture
print(img3.shape)

# removing shadow/noise from image which can be taken from phone camera

rgb_planes = cv2.split(img3)
result_planes = []
result_norm_planes = []
for plane in rgb_planes:
    dilated_img = cv2.dilate(plane, np.ones((10, 10), np.uint8))        #change the value of (10,10) to see different results
    bg_img = cv2.medianBlur(dilated_img, 21)
    diff_img = 255 - cv2.absdiff(plane, bg_img)
    norm_img = cv2.normalize(diff_img, None, alpha=0, beta=250, norm_type=cv2.NORM_MINMAX,
                                                 dtype=cv2.CV_8UC1)
    result_planes.append(diff_img)
    result_norm_planes.append(norm_img)

result = cv2.merge(result_planes)
result_norm = cv2.merge(result_norm_planes)
dst = cv2.fastNlMeansDenoisingColored(result_norm, None, 10, 10, 7, 11)             # removing noise from image

text = pytesseract.image_to_string(dst).upper().replace(" ", "")

date = str(re.findall(r"[\d]{1,4}[/-][\d]{1,4}[/-][\d]{1,4}", text)).replace("]", "").replace("[","").replace("'", "")
print(date)
number = str(re.findall(r"[0-9]{11,12}", text)).replace("]", "").replace("[","").replace("'", "")
print(number)
sex = str(re.findall(r"MALE|FEMALE", text)).replace("[","").replace("'", "").replace("]", "")
print(sex)

cv2.imshow('original',img3)
cv2.imshow('edited',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()


# crop_pic from ID card

gray = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = face_cascade.detectMultiScale(gray, 1.3, 7)
for (x, y, w, h) in faces:
    ix = 0
    cv2.rectangle(img3, (x, y), (x + w, y + h), (0, 255, 0), 2)
    roi_color = img3[y:y + h, x:x + w]
    #crop_pic = cv2.imwrite('croppic10.jpg', roi_color)
    crop_pic = cv2.imshow('croppicds', roi_color)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def pattern_matching(dic,lis):
    
    date_regex = r"[\d]{1,4}[/-][\d]{1,4}[/-][\d]{1,4}"
    aadhar_regex = r"[0-9]{4} [0-9]{4} [0-9]{4}"
    gender_regex = r"MALE|FEMALE|Female|Male"
    
    t = []

    for i in lis:
        if str(re.findall(date_regex, i)).replace("]", "").replace("[","").replace("'", "") != '':
            t.append(str(re.findall(date_regex, i)).replace("]", "").replace("[","").replace("'", ""))
    print(t)
    
    dic['DOB'] = t.copy()

    t.clear()

    for i in lis:
        if str(re.findall(gender_regex, i)).replace("]", "").replace("[","").replace("'", "") != '':
            t.append(str(re.findall(gender_regex, i)).replace("]", "").replace("[","").replace("'", ""))
    print(t)
    dic['Gender'] = t.copy()

    t.clear()

    for i in lis:
        if str(re.findall(aadhar_regex, i)).replace("]", "").replace("[","").replace("'", "") != '':
            t.append(str(re.findall(aadhar_regex, i)).replace("]", "").replace("[","").replace("'", ""))
    print(t)
    dic['Aadhar Number'] = t.copy()

    return dic
```

```python
from flair.data import Sentence,
from flair.models import SequenceTagger

# load tagger
tagger = SequenceTagger.load("flair/ner-english-ontonotes")
```

```python
# make example sentence
sentence = Sentence("On September 1st Malla won 1 dollar.")

# predict NER tags
tagger.predict(sentence)

# print sentence
print(sentence)

# print predicted NER spans
print('The following NER tags are found:')
# iterate over entities and print
for entity in sentence.get_spans('ner'):
    print(entity.text,entity.labels[0].value)
```
